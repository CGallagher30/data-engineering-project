```python
# List of urls to to start the scraping process

url_1='https://www.pgatour.com/stats/categories.ROTT_INQ.html'
url_2='https://www.pgatour.com/stats/categories.RAPP_INQ.html'
url_3='https://www.pgatour.com/stats/categories.RARG_INQ.html'
url_4='https://www.pgatour.com/stats/categories.RPUT_INQ.html'
url_5='https://www.pgatour.com/stats/categories.RSCR_INQ.html'
url_6='https://www.pgatour.com/stats/categories.RSTR_INQ.html'
url_7='https://www.pgatour.com/stats/categories.RMNY_INQ.html'
url_8='https://www.pgatour.com/stats/categories.RPTS_INQ.html'
url_9='https://www.pgatour.com/stats/categories.RFAF_INQ.html'

urls=[url_1,url_2,url_3,url_4,url_5,url_6,url_7,url_8,url_9]

# From here need to find the links to each stat page, starting with the above urls
```

```python
# Getting every link to each stat
links=[]
for url in urls:
    time.sleep(1+2*random.random())
    response=requests.get(url)
    time.sleep(1+2*random.random())
    print(response.status_code)
    html=response.text
    soup=bs(html)
    anchors=soup.find_all('a',attrs={'href':re.compile("^/stats/stat."),'aria-describedby':None})
    for anchor in anchors:
        href=anchor['href']
        stat=anchor.text
        print(stat)
        print(href)
        if href not in links:
            links.append(href)
```

```python
# Building a two master lists of links
eoff=[]
eon=[]
for link in links[:3]:
    url=base_url+link
    time.sleep(1+2*random.random())
    response=requests.get(url)
    time.sleep(1+2*random.random())
    print(response.status_code)
    html=response.text
    soup=bs(html)
    y=soup.find_all('option',attrs={'value':re.compile('y')})
    year_list=[tag['value'] for tag in y]
    for year in year_list:
        url=base_url+link
        url=url.replace('html','')+year+'.html'
        time.sleep(1+2*random.random())
        response=requests.get(url)
        time.sleep(1+2*random.random())
        print(response.status_code)
        html=response.text
        soup=bs(html)
        t=soup.find_all('option',attrs={'value':re.compile('t')})
        tournament_list=[tag['value'] for tag in t]
        for tournament in tournament_list:
            url=base_url+link
            url=url.replace('html','')+year+'.eoff.'+tournament+'.html'
            print(url)
            eoff.append(url)
            url=base_url+link
            url=url.replace('html','')+year+'.eon.'+tournament+'.html'
            print(url)
            eon.append(url)
```
